```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rotate(circle_in, T):
    angle_rad = np.radians(T)
    rotated_circle = []
    for x,y in circle_in:
        new_x = x*np.cos(angle_rad) - y*np.sin(angle_rad)
        new_y = x*np.sin(angle_rad) + y*np.cos(angle_rad)
        rotated_circle.append((new_x, new_y))

    return rotated_circle

#main
circle_in = [(6, 6), (10, 2), (14, 6), (10, 10)]

x, y = zip(*circle_in + [circle_in[0]])
plt.plot(x, y)
plt.title('original')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis('equal')
plt.show()

T = 20
rotated_circle = rotate(circle_in, T)
x, y = zip(*rotated_circle + [rotated_circle[0]])
plt.plot(x, y)
plt.title('original')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis('equal')
plt.show()
```
