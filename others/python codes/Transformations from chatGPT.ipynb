```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

def apply_transformation(points, matrix):
    homogenous_points = np.column_stack((points, np.ones(len(points))))
    transformed_points = np.dot(homogenous_points, matrix)
    return transformed_points[:, :-1]

def plot_polygon(ax, vertices, title):
    polygon = Polygon(vertices, closed=True, edgecolor='black')
    ax.add_patch(polygon)
    ax.set_title(title)
    ax.set_xlim([-10, 10])
    ax.set_ylim([-10, 10])
    ax.set_aspect('equal', adjustable='box')
    ax.grid(True)
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    ax.plot()

def main():
    # Original polygon vertices
    original_vertices = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])

    # Transformation matrices
    translation_matrix = np.array([[1, 0, 2],
                                   [0, 1, 3],
                                   [0, 0, 1]])

    shearing_matrix = np.array([[1, 2, 0],
                                [0, 1, 0],
                                [0, 0, 1]])

    reflection_matrix = np.array([[-1, 0, 0],
                                  [0, 1, 0],
                                  [0, 0, 1]])

    scaling_matrix = np.array([[2, 0, 0],
                               [0, 2, 0],
                               [0, 0, 1]])

    rotation_matrix = np.array([[np.cos(np.pi/4), -np.sin(np.pi/4), 0],
                                [np.sin(np.pi/4), np.cos(np.pi/4), 0],
                                [0, 0, 1]])

    # Apply transformations
    translated_vertices = apply_transformation(original_vertices, translation_matrix)
    sheared_vertices = apply_transformation(original_vertices, shearing_matrix)
    reflected_vertices = apply_transformation(original_vertices, reflection_matrix)
    scaled_vertices = apply_transformation(original_vertices, scaling_matrix)
    rotated_vertices = apply_transformation(original_vertices, rotation_matrix)

    # Plotting
    fig, axs = plt.subplots(2, 3, figsize=(15, 10))

    plot_polygon(axs[0, 0], original_vertices, "Original")
    plot_polygon(axs[0, 1], translated_vertices, "Translation")
    plot_polygon(axs[0, 2], sheared_vertices, "Shearing")
    plot_polygon(axs[1, 0], reflected_vertices, "Reflection")
    plot_polygon(axs[1, 1], scaled_vertices, "Scaling")
    plot_polygon(axs[1, 2], rotated_vertices, "Rotation")

    plt.show()

if __name__ == "__main__":
    main()
```
