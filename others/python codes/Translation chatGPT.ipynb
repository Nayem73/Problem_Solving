```python
import matplotlib.pyplot as plt
import numpy as np

def translate_polygon(polygon, tx, ty):
    # Translate each point of the polygon by (tx, ty)
    translated_polygon = [(x + tx, y + ty) for x, y in polygon]
    return translated_polygon

def plot_polygon(polygon, title):
    # Plot the original polygon
    x, y = zip(*polygon + [polygon[0]])  # Closing the polygon
    plt.plot(x, y, label='Original')

    # Add labels and title
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Set equal scaling for better visualization
    plt.axis('equal')

    # Display the legend
    plt.legend()

    # Show the plot
    plt.show()

def plot_translation_steps(polygon, translation_steps):
    for i, (tx, ty) in enumerate(translation_steps):
        translated_polygon = translate_polygon(polygon, tx, ty)

        # Plot the translated polygon
        x, y = zip(*translated_polygon + [translated_polygon[0]])  # Closing the polygon
        plt.plot(x, y, label=f'Step {i + 1}')

    # Add labels and title
    plt.title('Translation Steps')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Set equal scaling for better visualization
    plt.axis('equal')

    # Display the legend
    plt.legend()

    # Show the plot
    plt.show()

# Define the polygon (example: square)
polygon = [(0, 0), (1, 0), (1, 1), (0, 1)]

# Set translation steps (example: move right by 1 unit, up by 2 units)
translation_steps = [(1, 0), (0, 2)]

# Plot the original polygon
plot_polygon(polygon, 'Original Polygon')

# Plot the translation steps
plot_translation_steps(polygon, translation_steps)
```
