```python
import matplotlib.pyplot as plt
```

```python
def Bresenham(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    p = 2*dy - dx

    x, y = x1, y1
    line_x = [x]
    line_y = [y]
    while (x < x2):
        if p < 0:
            x += 1
            line_x.append(x)
            line_y.append(y)
            p = p + 2*dy

        else:
            x += 1
            y += 1
            line_x.append(x)
            line_y.append(y)
            p = p + 2*dy - 2*dx

    return line_x, line_y

#Example
x1, y1 = 9, 18
x2, y2 = 14, 22
line_x, line_y = Bresenham(x1,y1,x2,y2)

#plot
plt.plot(line_x, line_y, marker='o')
plt.title('Bresenham Line Drawing Algorithm')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
